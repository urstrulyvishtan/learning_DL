```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 14,6
```

```python
series = [3,10,12,13,12,10,12]
```

```python
def exponential_smoothing(series, alpha):
    result = [series[0]] # first value is same as series
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
        print (result)
    return result
```

```python
exponential_smoothing(series, 0.9)
```

```python
exponential_smoothing(series, 0.1)
```

```python
series_0_point_1 = exponential_smoothing(series, 0.1)
series_0_point_9 = exponential_smoothing(series, 0.9)
```

```python
plt.plot(series, color='black', linewidth=2, label='Original Series')
plt.plot(series_0_point_1, color='red', linewidth=2, label='0.1 Alpha Series')
plt.plot(series_0_point_9, color='blue', linewidth=2, label='0.9 Alpha Series')
plt.legend(loc='lower right')
```
