```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray
```

```python
# Define 4 by 4 original patch
orig_4_4 = np.array([[1, 3, 5,3 ], [6,2,0,8], [4,6,1,4], [2,8,0,3]])
print(orig_4_4)
```

```python
def subsample(x_in):
    x_out = np.zeros((int(np.ceil(x_in.shape[0] / 2)), int(np.ceil(x_in.shape[1] / 2))))
    # Subsampling routine
    for i in range(x_out.shape[0]):
        for j in range(x_out.shape[1]):
            x_out[i, j] = x_in[i * 2, j * 2]
    return x_out
```

```python
print("Original:")
print(orig_4_4)
print("Subsampled:")
print(subsample(orig_4_4))
```

```python
def maxpool(x_in):
    # Calculate the size of the output matrix
    x_out_height = int(np.floor(x_in.shape[0] / 2))
    x_out_width = int(np.floor(x_in.shape[1] / 2))
    
    x_out = np.zeros((x_out_height, x_out_width))
    
    # Perform max-pooling
    for i in range(x_out_height):
        for j in range(x_out_width):
            # Extract the 2x2 block
            block = x_in[2*i:2*i+2, 2*j:2*j+2]
            # Compute the maximum value in the block
            x_out[i, j] = np.max(block)
    
    return x_out
```

```python
print("Original:")
print(orig_4_4)
print("Maxpooled:")
print(maxpool(orig_4_4))
```

```python
# Let's see what Rick looks like:
data_maxpool = maxpool(data);

plt.figure(figsize=(5,5))
plt.imshow(data, cmap='gray')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(data_maxpool, cmap='gray')
plt.show()

data_maxpool2 = maxpool(data_maxpool)
plt.figure(figsize=(5,5))
plt.imshow(data_maxpool2, cmap='gray')
plt.show()

data_maxpool3 = maxpool(data_maxpool2)
plt.figure(figsize=(5,5))
plt.imshow(data_maxpool3, cmap='gray')
plt.show()
```

```python
def meanpool(x_in):
    # Calculate the size of the output matrix
    x_out_height = int(np.floor(x_in.shape[0] / 2))
    x_out_width = int(np.floor(x_in.shape[1] / 2))
    
    x_out = np.zeros((x_out_height, x_out_width))
    
    # Perform mean-pooling
    for i in range(x_out_height):
        for j in range(x_out_width):
            # Extract the 2x2 block
            block = x_in[2*i:2*i+2, 2*j:2*j+2]
            # Compute the mean value in the block
            x_out[i, j] = np.mean(block)
    
    return x_out
```

```python
print("Original:")
print(orig_4_4)
print("Meanpooled:")
print(meanpool(orig_4_4))
```

```python
# Let's see what Rick looks like:
data_meanpool = meanpool(data);

plt.figure(figsize=(5,5))
plt.imshow(data, cmap='gray')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(data_meanpool, cmap='gray')
plt.show()

data_meanpool2 = meanpool(data_maxpool)
plt.figure(figsize=(5,5))
plt.imshow(data_meanpool2, cmap='gray')
plt.show()

data_meanpool3 = meanpool(data_meanpool2)
plt.figure(figsize=(5,5))
plt.imshow(data_meanpool3, cmap='gray')
plt.show()
```

```python
# Define 2 by 2 original patch
orig_2_2 = np.array([[2, 4], [4,8]])
print(orig_2_2)
```

```python
def duplicate(x_in):
    # Calculate the size of the output matrix
    x_out = np.zeros((x_in.shape[0] * 2, x_in.shape[1] * 2))
    
    # Perform duplication
    for i in range(x_in.shape[0]):
        for j in range(x_in.shape[1]):
            # Assign the value to the 2x2 block in the output matrix
            x_out[2*i, 2*j] = x_in[i, j]
            x_out[2*i, 2*j + 1] = x_in[i, j]
            x_out[2*i + 1, 2*j] = x_in[i, j]
            x_out[2*i + 1, 2*j + 1] = x_in[i, j]
    
    return x_out
```

```python
print("Original:")
print(orig_2_2)
print("Duplicated:")
print(duplicate(orig_2_2))
```
