```python
import librosa
import librosa.display as disp
import scipy as sp
import IPython.display as ipd
import numpy as np
import matplotlib.pyplot as plt
```

```python
audio_path = 'audio/piano_c.wav'
ipd.Audio(audio_path)
```

```python
signal, sr = librosa.load(audio_path)
```

```python
plt.figure(figsize=(18, 8))
librosa.display.waveshow(signal, sr=sr, alpha=0.5)
plt.show()
```

```python
ft = sp.fft.fft(signal)
magnitude = np.absolute(ft)
frequency = np.linspace(0, sr, len(magnitude))
```

```python
plt.figure(figsize=(18, 8))
plt.plot(frequency[:5000], magnitude[:5000])
plt.xlabel('Frequency (Hz)')   
plt.ylabel('Magnitude')
plt.show()
```

```python
samples = range(0, len(signal))
t = librosa.samples_to_time(samples, sr=sr)

plt.figure(figsize=(18, 8))
plt.plot(t[10000:10400], signal[10000:10400])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
f = 523
phase = 0
phase_2 = 0.2

sin = 0.5 * np.sin(2 * np.pi * f * t - phase)
sin_2 = 0.5 * np.sin(2 * np.pi * f * t - phase_2)

plt.figure(figsize=(18, 8))
plt.plot(t[10000:10400], sin[10000:10400], color = "r")
plt.plot(t[10000:10400], sin_2[10000:10400], color = "y")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
f = 523
phase = 0.55
sin = 0.5 * np.sin(2 * np.pi * f * t - phase)

plt.figure(figsize=(18, 8))
plt.plot(t[10000:10400], signal[10000:10400])
plt.plot(t[10000:10400], sin[10000:10400], color = "r")

plt.fill_between(t[10000:10400], signal[10000:10400], sin[10000:10400], color = "y", alpha = 0.5)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
f = 1
t = np.linspace(0, 10, 10000)

sin = np.sin(2*np.pi * (f * t))
sin2 = np.sin(2*np.pi * (2*f * t))
sin3 = np.sin(2*np.pi * (3*f * t))

sum_signal = sin + sin2 + sin3

plt.figure(figsize=(15, 10))

plt.subplot(4, 1, 1)
plt.plot(t, sum_signal, color="r")

plt.subplot(4, 1, 2)
plt.plot(t, sin)

plt.subplot(4, 1, 3)
plt.plot(t, sin2)

plt.subplot(4, 1, 4)
plt.plot(t, sin3)

plt.show()
```
